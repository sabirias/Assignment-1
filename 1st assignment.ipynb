```python
6) Name the few blocks of code used to define in python Language?

ANS:- In Python, code blocks or suites are defined by indentation. Indentation is a crucial part of the Python syntax and
      is used to indicate the grouping of statements. Here are a few types of code blocks commonly used in Python:

      
1. Function Definition:
   
    def my_function():
        # Code block for the function
        print("Hello, World!")

        The indented block under `def my_function():` is the suite of code belonging to the function.


2. Conditional Statements (if, elif, else):

    if condition:
        # Code block for true condition
        print("Condition is true")
    else:
        # Code block for false condition
        print("Condition is false")

   Both the `if` and `else` blocks are defined by indentation.


3. Loops (for, while):

    for item in iterable:
        # Code block for the loop
        print(item)

   Similarly, the indented block under the `for` statement is the suite of code that gets executed in each iteration.

4. Class Definitions:

    class MyClass:
        # Code block for the class
        def __init__(self, x):
            self.x = x

   The code block under `class MyClass:` defines the suite of code belonging to the class.

5. Exception Handling (try, except):

    try:
        # Code block that may raise an exception
        result = 10 / 0
    except ZeroDivisionError:
        # Code block for handling the exception
        print("Cannot divide by zero.")

   Both the `try` and `except` blocks are defined by indentation.


In each of these cases, the code blocks are determined by the level of indentation. 
Consistent and proper indentation is crucial in Python to convey the structure of the code accurately.
```

```python
8) Mention function which can help us to find the version of python that we are currently working on?

ANS:- To find the version of Python that you are currently working with, you can use the sys module along with its version attribute.
      Here's an example:
    
import sys

print("Python version")
print(sys.version)
print("Version info.")
print(sys.version_info)

The sys.version attribute returns a string containing the version number of the Python interpreter, and sys.
version_info returns a tuple with the major,minor, micro, release level, and serial values of the Python version.
This information can be helpful in determining the exact version of Python you are using.
```

```python
9) Pyhton support the creation of anonymous function at runtime, using a construct called _ _ _ _ _ _ _ _ _ _

ANS:- In Python, the construct used for creating anonymous functions at runtime is called `lambda`. The `lambda` keyword allows you to
create small,unnamed functions without explicitly defining them using the `def` keyword.

The syntax for a lambda function is as follows:

lambda arguments: expression

Here's a simple example of a lambda function:

# Anonymous function that adds two numbers
add = lambda x, y: x + y

result = add(3, 5)
print(result)  # Output: 8

Lambda functions are often used for short, simple operations, and they are particularly useful when you need to pass a small function
as an argument to higher-order functions like `map()`, `filter()`, or `sorted()`.
```

```python
11) Mention a few built-in function in python?

ANS:- Certainly! Python comes with a rich set of built-in functions that provide a wide range of functionality.
      Here are a few commonly used built-in functions in Python:

1. print(): Outputs text or variables to the console.
   print("Hello, World!")

2. len(): Returns the length of an object (e.g., a string, list, or tuple).
   my_list = [1, 2, 3, 4, 5]
   length = len(my_list)

3. input(): Reads a line from the console input and returns it as a string.
   user_input = input("Enter something: ")

4. type(): Returns the type of an object.
   x = 10
   print(type(x))  # Output: <class 'int'>

5. range(): Generates a sequence of numbers.
   numbers = list(range(5))

6. sum(): Returns the sum of elements in an iterable.
   my_list = [1, 2, 3, 4, 5]
   total = sum(my_list)

7. max() and min(): Return the maximum or minimum value in an iterable.
   my_list = [1, 2, 3, 4, 5]
   max_value = max(my_list)
   min_value = min(my_list)

8. abs(): Returns the absolute value of a number.
   x = -10
   absolute_value = abs(x)

9. round(): Rounds a floating-point number to the nearest integer.
   x = 3.14159
   rounded_value = round(x, 2)  # Rounds to 2 decimal places

10. sorted(): Returns a sorted list from the elements of an iterable.
    my_list = [4, 2, 8, 1, 6]
    sorted_list = sorted(my_list)

These are just a few examples. Python has many more built-in functions that cover a wide range of tasks,
from mathematical operations to string manipulation, file handling, and more.
```

14) How is memory managed in python?

ANS:- Memory management in Python is a combination of reference counting and garbage collection. Reference counting is a technique that tracks how many               references are to an object. When the reference count reaches zero, the object is deallocated. Garbage collection is a process that automatically               reclaims memory that is no longer being used.
